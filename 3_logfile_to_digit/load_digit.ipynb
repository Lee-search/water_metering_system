## Load Digit from logflies dataset

```python
import pandas as pd
f_name = 'data_2022-12-19.csv'
df = pd.read_csv('data/' + f_name,
                 sep=',',
                 names=['time', 'main', 'digit', 'NaN_0', 'NaN_1', 'NaN_2', 'NaN_3', 'str', 'digit_1000', 'digit_100', 'digit_10', 'digit_1'])
```

```python
df
```

```python
times = []
digits = []

for i in range(len(df)):
    t = df['time'][i]
    d = df['digit'][i]
    # 로그에서 'Rate too high' 예외처리 후 digit 4자리 모두 제대로 읽힌 출력물만 저장
    if df['str'][i][:4] == 'Neg.' and 'N' not in d:
        times.append(t)
        digits.append(d)
```

```python
df_clear = pd.DataFrame({'time': times,'digit': digits})
df_clear.to_csv('output/' + f_name)
```
